```python
# read the first CSV file
df1 = pd.read_csv('/content/drive/MyDrive/final_year_project/data/Cleaned_Data/Bangalore_Restaurants.csv')
```

```python
df1 = df1.drop(["Dining_Review_Count"], axis=1)
```

```python

# read the second CSV file
df4 = pd.read_csv('/content/drive/MyDrive/final_year_project/data/Cleaned_Data/zomato_restaurants_in_India.csv')
```

```python
df4.head()
```

```python
df4['establishment']=df4['establishment'].str[2:-2]
df4 = df4.drop(['zipcode'], axis=1)
df4.head()
```

```python
df4['cuisines']=df4['cuisines']+" - "+df4['establishment']
df4 = df4.drop(['city','locality','rating_text','establishment'], axis=1)
```

```python
df4.head()
```

```python
df4=df4.rename(columns={'name': 'Restaurant_Name', 'cuisines': 'Cuisines','url':'Website','address':'Address','locality_verbose':'Locality','latitude':'Latitude','longitude':'Longitude','average_cost_for_two':'Pricing_for_2','aggregate_rating':'Dining_Rating','timings':'Timings','highlights':'Highlights','votes':'Votes','delivery':'Delivery','takeaway':'Takeaway'})
df4.head()
```

```python
df1=df1.rename(columns={'Category':'Cuisines'})
```

```python
df1.head()
```

```python
df5=pd.concat([df1,df4])
df5.head()
```
